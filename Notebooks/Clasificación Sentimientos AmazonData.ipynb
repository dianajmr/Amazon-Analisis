```python
import nltk
import pandas as pd
from nltk.corpus import stopwords
from afinn import Afinn
```

```python
# Descargar stopwords en inglés de nltk
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
```

```python
nltk.download('punkt')
```

```python
# Cargar el dataset
df = pd.read_csv("result/AmazonData_Categories_DistrilBERT.csv")
```

```python
# Tokenización y preprocesamiento
afinn = Afinn()
polarities = []

for text in df['Text']:
    tokens = nltk.word_tokenize(text.lower())  
    filtered_tokens = [word for word in tokens if word.isalnum() and word not in stop_words]  
    polarity = sum(afinn.score(token) for token in filtered_tokens)  
    polarity = max(-5, min(5, polarity))
    polarities.append(polarity)
```

```python
# Clasificación de sentimientos
df['Polarity'] = polarities
df['Sentiment'] = df['Polarity'].apply(lambda x: 'Positive' if x > 0 else ('Negative' if x < 0 else 'Neutral'))
```

```python
# Verificación de resultados
df.groupby(['Category', 'Sentiment'],as_index=False).size()
```
