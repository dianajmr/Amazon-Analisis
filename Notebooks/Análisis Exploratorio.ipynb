```python
title = ' '.join(info['Title'])
wordcloud = WordCloud(width=800, height=600, background_color='white').generate(title)
plt.figure(figsize=(20, 15))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Temtica de los ttulos', fontdict={'fontsize': 20})
plt.show()
```

```python
text = ' '.join(info['Text'])
wordcloud = WordCloud(width=800, height=600, background_color='white').generate(text)
plt.figure(figsize=(25, 12))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Temtica de opiniones', fontdict={'fontsize': 20})
plt.show()
```
